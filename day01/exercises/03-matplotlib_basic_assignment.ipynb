## Exercise 1
**Read Total profit of all months and show it using a line plot**

Total profit data provided for each month. Generated line plot must include the following properties:
- X label name = Month Number
- Y label name = Total profit
The line plot graph should look like this.

**Expected output**

<img src="../data/Company_profit_per_month.png" alt="Company_profit_per_month" style="width: 500px;"/>


```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("../data/company_sales_data.csv")


axis1=data["total_profit"]
axis0=data["month_number"]
plt.plot(axis0,axis1,scalex=True)

plt.xlabel("Month number")
plt.ylabel("Profit in dollar")
plt.title("Company profit per month")
```

## Exercise 2
**Get total profit of all months and show line plot with the following Style properties**
- Line Style dotted and Line-color should be red
- Show legend at the lower right location.
- X label name = Month Number
- Y label name = Sold units number
- Add a circle marker.
- Line marker color as read
- Line width should be 3

**Expected output**

<img src="../data/company_sales_data_of_last_year.png" alt="company_sales_data_of_last_year" style="width: 700px;"/>


```python
plt.figure(figsize=(10,5))
plt.plot(data["month_number"],data["total_profit"],color='r',marker='o',mfc='black',linestyle="dotted",linewidth=3,label="Profit data last year")
plt.legend(loc=4)
plt.locator_params('x',nbins = 20)
plt.xlabel("Month number")
plt.ylabel("solid units number")
plt.title("Company sales data of last year")
```

## Exercise 3
**Read all product sales data and show it  using a multiline plot**
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).
**Expected output**

<img src="../data/sales_data.png" alt="sales_data" style="width: 500px;"/>


```python
plt.figure(figsize=(10,5))
plt.plot(data.month_number,data.facecream, linestyle='-', label="Face cream Sales Data", marker='o' )
plt.plot(data.month_number,data.facewash, linestyle='-', label="Face Wash Sales Data", marker='o' )
plt.plot(data.month_number,data.toothpaste, linestyle='-', label="ToothPaste Sales Data", marker='o' )
plt.plot(data.month_number,data.bathingsoap, linestyle='-', label="Bathingshop Sales Data", marker='o' )
plt.plot(data.month_number,data.moisturizer, linestyle='-', label="Moisturizer Sales Data", marker='o' )
plt.plot(data.month_number,data.shampoo, linestyle='-', label="Shampoo Sales Data", marker='o' )
plt.xlabel("Months Number")
plt.ylabel("Sales units in number")
plt.legend(loc=2)
plt.ylim(1000)
plt.show()
```

## Exercise 7
**Read the total profit of each month and show it using the histogram to see the most common profit ranges**

**Expected output**

## Exercise 8
**Calculate total sale data for last year for each product and show it using a Pie chart**

*Note: In Pie chart display Number of units sold per year for each product in percentage.*

**Expected output**

<img src="../data/sales_data_pi_chart.png" alt="sales_data_pi_chart" style="width: 700px;"/>


```python

total_sale = [sum(data.facecream), sum(data.facewash), sum(data.toothpaste), sum(data.bathingsoap), sum(data.shampoo), sum(data.moisturizer)]
label = ['Facecream', 'Facewash', 'Toothpaste', 'Bathingsoap','Shampoo','Moisturizer']

plt.pie(total_sale , autopct='%1.1f%%', labels=label)
plt.legend(loc='center left', bbox_to_anchor=(1.5, 0.5))
plt.show()
```

## Exercise 9
**Read Bathing soap facewash of all months and display it using the Subplot**

**Expected output**

<img src="../data/Sales_data_of_a_facewash.png" alt="Sales_data_of_a_facewash" style="width: 700px;"/>


```python

s,subplt = plt.subplots(2, sharex=True)
subplt[0].plot(data.month_number, data.bathingsoap, label = 'Bathingsoap Sales Data', color='k', marker='o', linewidth=3)
subplt[0].set_title('Sales data of  a Bathingsoap')
subplt[1].plot(data.month_number, data.facewash, label = 'Face Wash Sales Data', color='r', marker='o', linewidth=3)
subplt[1].set_title('Sales data of  a facewash')


plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.show()
```

## Exercise 10
**Read all product sales data and show it using the stack plot**

**Expected output**

<img src="../data/product_sales_stack_plot.png" alt="product_sales_stack_plot" style="width: 700px;"/>


```python
facc  =data.loc[:,'facecream']
facw   = data.loc[:,'facewash']
totp = data.loc[:,'toothpaste']
bats   = data.loc[:,'bathingsoap']
shp  = data.loc[:,'shampoo']
mois = data.loc[:,'moisturizer']
months = data.loc[:,'month_number']

plt.figure(figsize=(10,5))
plt.plot([],[],color='m', label='face Cream')
plt.plot([],[],color='c', label='Face wash')
plt.plot([],[],color='r', label='Tooth paste')
plt.plot([],[],color='k', label='Bathing soap')
plt.plot([],[],color='g', label='Shampoo')
plt.plot([],[],color='y', label='Moisturizer')

plt.stackplot(months, facc, facw, totp, bats, shp, mois, colors=['m','c','r','k','g','y'])
plt.legend(loc="upper left")
plt.xlabel('Month Number')
plt.ylabel('Sale units')
plt.show()
```
